```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def normal_histogram(img):    
    
    pixelmap = list(img.getdata())
    elements = list(set(pixelmap))
    
    frequency = np.zeros((len(elements),),dtype=float)
    
    for i in range(len(elements)):   
        val = pixelmap.count(elements[i])
        frequency[i] = val
    
    prob=frequency/(len(elements))
    return elements, frequency,prob
```

```python
image = Image.open('cameraman.tif')
new_img = image.copy()
elements,frequency,prob = normal_histogram(new_img)
```

```python
class element:  
    def __init__(self, prob, marked,label,code):  
        self.prob = prob  
        self.marked = marked 
        self.label=label
        self.code=code
    def __str__(self):
        return str(self.label)+"-"+str(self.marked)+"-"+str(self.prob)+"-"+str(self.code)
```

```python
elements=['a2','a6','a1','a4','a3','a5']
prob=[0.4,0.3,0.1,0.1,0.06,0.04]
```

```python
def generateList(newlist):
    mList=[]
    for k in range (0,len(newlist)-1):
        if(k!=len(newlist)-2):
            mList.append( element(newlist[k].prob,False,None,'') ) 
        else:
            mList.append( element(round(newlist[k].prob + newlist[k+1].prob,5) ,True,None,'')) 

    # after we come up mark and sort        
    mList.sort(key=lambda x: x.prob, reverse=True)
    return mList
```

```python
def func1(after,before):
    for k in range (0,len(after)):
        if(after[k].marked==True):
            code=after[k].code
    for i in range (0,len(before)):
        if(i>0 and before[i].prob==before[i-1].prob):
            rep+=1
        else:
            rep=0  
        for j in range(0,len(after)):
            if(before[i].prob == after[j].prob and after[j].marked==False and i <len(before)-2 ):
                before[i].code=after[j+rep].code
                break
            # abl el akhir
            elif(i==len(before)-2):
                before[i].code=code +'0'
            elif(i==len(before)-1):
                before[i].code=code +'1'
                
                
    
```

```python
def huffman(elements,prob):
    largeList=[]
    mainList = []  
    for j in range (0,len(elements)):
        mainList.append( element(prob[j],False,elements[j],None)) 
        
    newlist = sorted(mainList, key=lambda x: x.prob, reverse=True)    
    largeList.append(newlist)
    while(True):
        mList = generateList(newlist)
        largeList.append(mList)
        newlist= mList
        if len(mList)<=2 :
            break
    #here
    index= len(largeList)-1
    while(index>0):
        if(index==len(largeList)-1):
            largeList[index][0].code='0'
            largeList[index][1].code='1'
            func1(largeList[index],largeList[index-1])
        else:
            func1(largeList[index],largeList[index-1])
        index-=1    
    return largeList  
```

```python
final=huffman(elements,prob)
```

```python
for i in range (0,len(final[0])):
    print(str(final[0][i].label) + "-" + str(final[0][i].prob)+"-"+str(final[0][i].code))
```
